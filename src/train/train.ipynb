```python
import torch
from torch.utils.data import DataLoader
from src.train.load import MasterDataset, _worker_init_fn
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

device
```

```python
def to_tensors(sample):
    x = {
        "gz": torch.from_numpy(sample["gz"]),                         # [N_recv]
        "receiver_locations": torch.from_numpy(sample["receiver_locations"]), # [N_recv, 3]
        "ind_active": torch.from_numpy(sample["ind_active"].astype("float32")),  # [mesh.nC]
    }
    y = torch.from_numpy(sample["true_model"])  # [n_active]
    return x, y

ds = MasterDataset("data/master.h5", transform=to_tensors)
```

```python

def _worker_init_fn(_): ds.close()

dl = DataLoader(
    ds, batch_size=8, shuffle=True,
    num_workers=4, pin_memory=True,
    worker_init_fn=_worker_init_fn,
    persistent_workers=True
)
```
